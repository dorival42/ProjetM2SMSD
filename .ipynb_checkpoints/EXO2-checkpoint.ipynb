```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the dataset
data = pd.read_csv("Frogs_MFCCs.csv", delimiter=",")

# Extract numeric variables, assuming the last 4 columns are non-numeric
# The first part of the data up to the second last 4 columns is numeric
numerique_data = data.iloc[:, :-4]  # Excluding the last 4 columns

data
```

```python
numerique_data 
```

## **1. Can we reduce our number of numeric variables?**
  
Yes, applying Principal Component Analysis (PCA) shows that the dataset's dimensionality can be reduced. By retaining 95% of the variance in the data, we only need 12 principal components instead of the original 22 numeric variables.

```python
# Step 1: Data Preprocessing (Normalization)
scaler = StandardScaler()
numerique_data_scaled = scaler.fit_transform(numerique_data)

# scaler = StandardScaler()
# X_scaled = scaler.fit_transform(train_test(data, normalize=True)[0])  # Normalize the features 
# y = train_test(data, normalize=True)[2]  

# Step 2: Apply PCA to the normalized data
pca = PCA()
X_pca = pca.fit_transform(numerique_data_scaled)

# Step 3: Visualize the cumulative explained variance
explained_variance_ratio = np.cumsum(pca.explained_variance_ratio_)
plt.figure(figsize=(10, 6))
plt.plot(range(1, numerique_data_scaled.shape[1] + 1), explained_variance_ratio, marker='o', linestyle='--', color='b')
plt.title("Cumulative Explained Variance by Principal Components")
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.grid()
plt.show()

# Step 4: Determine the number of components needed for different explained variances
seuils = [0.5, 0.6, 0.7, 0.8, 0.9, 0.95]
axes_retention = [np.argmax(explained_variance_ratio >= seuil) + 1 for seuil in seuils]
print(f"Components for different explained variances: {dict(zip(seuils, axes_retention))}")

# Find the number of components for 90% variance explained
seuil_95 = 0.95
n_components_95 = np.argmax(explained_variance_ratio >= seuil_95) + 1
print("--------------------------------------------------------------------------")
print(f"Number of components needed for 95% explained variance: {n_components_95}")


# Step 5: Apply PCA for dimensionality reduction
pca_reduced = PCA(n_components=n_components_95)  # Here we reduce the data to 12 dimensions 
numerique_data_pca = pca_reduced.fit_transform(numerique_data_scaled)

# Print the reduced data
print("--------------------------------------------------------------------------")
print("Dimensionality reduced data:")
print(numerique_data_pca)

# Print the explained variance ratio for each principal component
print("--------------------------------------------------------------------------")
print("Explained variance ratio:")
print(pca_reduced.explained_variance_ratio_)

# Step 6: Visualize the explained variance for the first 12 components
plt.figure(figsize=(10, 6))
plt.bar(range(1, 13), pca_reduced.explained_variance_ratio_, color='b', alpha=0.7)
plt.title("Explained Variance of the First 12 Principal Components")
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.show()
```


## **3. Characterize the qualitative variables based on the above analysis.**
     

### Dimensionality Reduction (PCA):

### Why PCA?
Principal Component Analysis (PCA) transforms the original numeric variables $X_1, X_2, \dots, X_{22}$ into a smaller number of uncorrelated principal components $Z_1, Z_2, \dots, Z_{12}\$, where each principal component is a weighted sum of the original variables. Mathematically, the transformation can be written as:

$$
Z_k = \sum_{i=1}^{22} w_{ki} X_i
$$

Where:
- $Z_k$ represents the \(k\)-th principal component.
- $w_{ki}$ is the weight (or loading) associated with each original variable $X_i$ in the $k$-th principal component.
- $X_i$ represents the original variables (in this case, the 22 numeric variables).

The weights \(w_{ki}\) indicate how much each original variable contributes to a given principal component \(Z_k\).

---

### Reduction Achieved:
- **From 22 numeric variables**: PCA reduces the dataset to **12 principal components** while retaining **95% of the variance** in the data.
- This dimensionality reduction simplifies the analysis by reducing the number of variables while still capturing the majority of the data's variability.
- Reducing dimensions helps to mitigate redundancy, as some of the original numeric variables were highly correlated, indicating overlapping information.

---

### Relationships Between Numeric Variables:
- The **PCA loadings matrix** (found in the PCA output) shows how each numeric variable contributes to each principal component.
  - For example:
    - **PC1** has significant contributions from `MFCCs_13`, `MFCCs_17`, and `MFCCs_22`.
    - **PC2** is influenced by `MFCCs_3`, `MFCCs_2`, and `MFCCs_17`.
  
This matrix allows us to understand the relationships between the original variables and the principal components, identifying which variables are important for each component.
