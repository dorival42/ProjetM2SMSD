```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist, squareform

# Charger les données pour examiner leur structure
file_path = 'Frogs_MFCCs.csv'
data = pd.read_csv(file_path)

# Afficher les premières lignes et les informations sur les colonnes
data_info = {
    "head": data.head(),
    "info": data.info(),
    "description": data.describe(include='all')
}

data_info
```

```python


# Calculer la matrice de corrélation pour les colonnes MFCCs
mfcc_columns = [col for col in data.columns if col.startswith('MFCCs_')]
correlation_matrix = data[mfcc_columns].corr()

# Visualiser la matrice de corrélation
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', cbar=True)
plt.title('Matrice de corrélation des MFCCs')
plt.show()

# Identifier les paires de colonnes fortement corrélées (corrélation > 0.9 ou < -0.9)
strongly_correlated_pairs = [
    (i, j, correlation_matrix.loc[i, j])
    for i in correlation_matrix.columns
    for j in correlation_matrix.columns
    if i != j and abs(correlation_matrix.loc[i, j]) > 0.9
]

strongly_correlated_pairs
```

```python
# Créer une copie de la matrice de corrélation pour marquer les corrélations fortes
highlighted_correlation_matrix = correlation_matrix.applymap(
    lambda x: f'{x:.2f}' if abs(x) > 0.5 else ''
)

# Visualiser la matrice de corrélation avec des annotations
plt.figure(figsize=(12, 10))
sns.heatmap(
    correlation_matrix, 
    annot=highlighted_correlation_matrix, 
    fmt='', 
    cmap='coolwarm', 
    cbar=True, 
    annot_kws={"size": 8, "color": "black"}
)
plt.title('Matrice de corrélation des MFCCs (valeurs fortes annotées)')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardiser les données pour PCA
scaler = StandardScaler()
mfcc_scaled = scaler.fit_transform(data[mfcc_columns])

# Appliquer PCA
pca = PCA()
pca_result = pca.fit_transform(mfcc_scaled)

# Variance expliquée par chaque composante principale
explained_variance = pca.explained_variance_ratio_

# Visualiser la variance expliquée cumulative
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance.cumsum(), marker='o', linestyle='--', color='b')
plt.title('Variance expliquée cumulative par PCA')
plt.xlabel('Nombre de composantes principales')
plt.ylabel('Variance expliquée cumulative')
plt.grid()
plt.show()

# Nombre de composantes nécessaires pour expliquer au moins 95% de la variance
n_components_95 = next(i for i, cumulative_variance in enumerate(explained_variance.cumsum(), start=1) if cumulative_variance >= 0.95)
n_components_95
```

```python
# Extraire les valeurs propres, les variances expliquées, et les composantes
eigenvalues = pca.explained_variance_
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = explained_variance_ratio.cumsum()
components = pca.components_

# Créer les tableaux correspondants
eigenvalues_table = pd.DataFrame({
    "Composante": [f"PC{i+1}" for i in range(len(eigenvalues))],
    "Valeur Propre": eigenvalues
})

variance_table = pd.DataFrame({
    "Composante": [f"PC{i+1}" for i in range(len(explained_variance_ratio))],
    "Variance Expliquée": explained_variance_ratio,
    "Variance Cumulative": cumulative_variance_ratio
})

loadings_table = pd.DataFrame(components.T, columns=[f"PC{i+1}" for i in range(len(components))], index=mfcc_columns)

eigenvalues_table
```

```python
import numpy as np

# Calculer les contributions des variables aux deux premières composantes principales
pc1, pc2 = components[0], components[1]

# Calculer la corrélation entre les variables et les composantes principales
circle_coords = pd.DataFrame({
    'PC1': pc1 * np.sqrt(eigenvalues[0]),
    'PC2': pc2 * np.sqrt(eigenvalues[1]),
}, index=mfcc_columns)

# Tracer le cercle des corrélations
plt.figure(figsize=(8, 8))
plt.axhline(0, color='grey', linestyle='--', linewidth=0.7)
plt.axvline(0, color='grey', linestyle='--', linewidth=0.7)

# Tracer les vecteurs
for i in circle_coords.index:
    plt.arrow(0, 0, circle_coords.loc[i, 'PC1'], circle_coords.loc[i, 'PC2'],
              color='b', alpha=0.7, head_width=0.03, head_length=0.05)
    plt.text(circle_coords.loc[i, 'PC1'] * 1.1, circle_coords.loc[i, 'PC2'] * 1.1,
             i, color='black', ha='center', va='center')

# Ajuster le graphique
plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)
plt.title('Cercle des corrélations (PC1 vs PC2)')
plt.xlabel('Composante principale 1 (PC1)')
plt.ylabel('Composante principale 2 (PC2)')
plt.grid()
plt.show()
```

```python
# Recommencer avec une circonférence tracée pour le cercle des corrélations
plt.figure(figsize=(8, 8))
plt.axhline(0, color='grey', linestyle='--', linewidth=0.7)
plt.axvline(0, color='grey', linestyle='--', linewidth=0.7)

# Dessiner la circonférence
circle = plt.Circle((0, 0), 1, color='black', fill=False, linestyle='--', linewidth=1.2)
plt.gca().add_artist(circle)

# Tracer les vecteurs des variables
for i in circle_coords.index:
    plt.arrow(0, 0, circle_coords.loc[i, 'PC1'], circle_coords.loc[i, 'PC2'],
              color='b', alpha=0.7, head_width=0.03, head_length=0.05)
    plt.text(circle_coords.loc[i, 'PC1'] * 1.1, circle_coords.loc[i, 'PC2'] * 1.1,
             i, color='black', ha='center', va='center')

# Ajuster les limites et le titre
plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)
plt.title('Cercle des corrélations (avec circonférence)')
plt.xlabel('Composante principale 1 (PC1)')
plt.ylabel('Composante principale 2 (PC2)')
plt.grid()
plt.show()
```

```python

# Charger les données
data = pd.read_csv("Frogs_MFCCs.csv")  # Remplacez par votre chemin
numerical_columns = [col for col in data.columns if col.startswith("MFCCs_")]

# Calcul de la matrice de corrélation
correlation_matrix = data[numerical_columns].corr()

# Conversion en matrice de distance (1 - |correlation|)
distance_matrix = 1 - correlation_matrix.abs()

# Clustering hiérarchique
linkage_matrix = linkage(squareform(distance_matrix), method="ward")

# Tracer le dendrogramme
plt.figure(figsize=(12, 8))
dendrogram(linkage_matrix, labels=numerical_columns, leaf_rotation=90, leaf_font_size=10)
plt.title("Dendrogramme - Clustering Hiérarchique")
plt.xlabel("Variables")
plt.ylabel("Distance")
plt.show()
```
